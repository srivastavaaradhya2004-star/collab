```python


# Dictionary of menu items and prices
menu = {
    "cold tea": 70,
    "chowmein": 56,
    "burger": 80,
    "pizza": 150,
    "momos": 90,
    "coffee": 60,
    "sandwich": 75,
    "pasta": 120
}

print(" WELCOME TO OUR RESTAURANT\n")

print(" MENU ")
for item, price in menu.items():
    print(f"{item} : ₹{price}")

order_amount = 0

while True:
    order = input("\nEnter the item you want to order: ").lower()

    if order in menu:
        order_amount += menu[order]
        print(f"{order} added successfully!")
    else:
        print("Sorry! This item is not available in the menu.")
        print("Item removed.")

    choice = input("Do you want to order another item? (yes/no): ").lower()

    if choice != "yes":
        break

print("\n BILL")
print("Your Total Bill Amount = ₹", order_amount)
print("Thank You! Visit Again ")
```
